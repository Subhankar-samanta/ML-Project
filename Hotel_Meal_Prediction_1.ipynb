##  Q1) Most popular meal Plan in 2018

```python
meal_plan_2018= df[df['arrival_year']==2018]['type_of_meal_plan'].value_counts().idxmax()
print("Most popular meal Plan in 2018:", meal_plan_2018)
```

##  Q2) What is the average lead time between a guest booking and the actual trip date

```python
average_lead_time = df['lead_time'].mean()
print ("Average Lead Time:", average_lead_time)
```

## Q3) Showcase a distribution between Weekend booking Weekdays Booking

```python
weekday_bookings= df[df['no_of_week_nights']==0]['arrival_date'].value_counts().sort_index()
weekend_bookings= df[df['no_of_weekend_nights'] > 0]['arrival_date'].value_counts().sort_index()
print("Weekday bookings distribution:")
print(weekday_bookings)
print("\nweekend booking distribution:")
print(weekend_bookings)
```
